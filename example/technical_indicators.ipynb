```python
from microcrypta.coin import Coin
from microcrypta.account import correlation_matrix, Account
from microcrypta.coinmarket import CoinMarket
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from microcrypta.bsh import BSH
from microcrypta.backtest import Backtest
import microcrypta.config as config
```

```python
corr_coins = ['BTC', 'ETH','SHIB','ADA']
correlation_matrix(corr_coins,time_frame='4h')
```

```python
coin = Coin('REQ')
usd = Coin('BTC')
coin_usd = CoinMarket(coin,usd,time_frame = "1h")
bshs = BSH(coin_usd)
```

```python
coin_usd.technical_summary()
```

```python
bt_macd = Backtest(coin_usd,bshs.macd(),signal_name = "MACD")
position_macd,coin_macd,percentage_macd = bt_macd.backtest(plot=False)

bt_rsi = Backtest(coin_usd,bshs.rsi(window=14),signal_name = "RSI")
position_rsi,coin_rsi,percentage_rsi = bt_rsi.backtest(plot=False)

bt_ma = Backtest(coin_usd,bshs.moving_average(),signal_name = "Moving Average")
position_ma,coin_ma,percentage_ma = bt_ma.backtest(plot=False)

bt_ao = Backtest(coin_usd,bshs.awesome_oscillator(),signal_name = "Awesome Oscillator")
position_ma,coin_ma,percentage_ma = bt_ao.backtest(plot=False)

bt_bb = Backtest(coin_usd,bshs.bollinger_bands(),signal_name = "Bollringer Band")
position_bb,coin_bb,percentage_bb = bt_bb.backtest(plot=False)

bt_so = Backtest(coin_usd,bshs.stochastic_oscillator(),signal_name = "Stochastic Oscillator")
position_so,coin_so,percentage_so = bt_so.backtest(plot=False)

bt_ai = Backtest(coin_usd,bshs.aroon(),signal_name = "Aroon Oscillator")
position_ai,coin_ai,percentage_ai = bt_ai.backtest(plot=False)
```

```python
bt_macd.plot_signal()
bt_rsi.plot_signal()
bt_ma.plot_signal()
bt_ao.plot_signal()
bt_bb.plot_signal()
bt_so.plot_signal()
bt_ai.plot_signal()
```
